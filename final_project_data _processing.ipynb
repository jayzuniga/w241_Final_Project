```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('primary_data.csv')
df = df[2:].copy().reset_index(drop=True)

mturk_df = pd.read_csv('primary_mturk_data.csv')
```

```python
df.columns = ['start_dt_tm',
              'end_dt_tm',
              'status_response_type',
              'ip_address',
              'progress', 
              'duration_seconds',
              'finished',
              'recorded_dt_tm',
              'response_id',
              'recipient_last_name', 
              'recipient_first_name',
              'recipient_email', 
              'external_reference',
              'location_latitude', 
              'location_longitude',
              'distributional_channel',
              'user_language', 
              'gender',
              'geder_text',
              'location',
              'political_identification',
              'political_identification_text',
              'age', 
              'time_first_click_article_control',
              'time_last_click_article_control',
              'time_submit_article_control',
              'click_count_article_control', 
              'time_first_click_article_treatment',
              'time_last_click_article_treatment',
              'time_submit_article_treatment',
              'click_count_article_treatment',
              'point_of_article', 
              'article_factual', 
              'article_effectiveness', 
              'support_tb', 
              'recommend_article', 
              'time_first_click_survey',
              'time_last_click_survey',
              'time_submit_survey',
              'click_count_survey',
              'group',
              'mturk_code']

df.mturk_code.fillna(0, inplace=True)
df['mturk_code'] = df.mturk_code.astype('str')

df['time_first_click_article'] = pd.concat([df.time_first_click_article_control.dropna(),
                                            df.time_first_click_article_treatment.dropna()]).reindex_like(df)

df['time_last_click_article'] = pd.concat([df.time_last_click_article_control.dropna(), 
                                                   df.time_last_click_article_treatment.dropna()]).reindex_like(df)

df['time_submit_article'] = pd.concat([df.time_submit_article_control.dropna(), 
                                       df.time_submit_article_treatment.dropna()]).reindex_like(df)

df['click_count_article'] = pd.concat([df.click_count_article_control.dropna(),
                                       df.click_count_article_treatment.dropna()]).reindex_like(df)

df.drop(['time_first_click_article_control',
         'time_first_click_article_treatment', 
         'time_last_click_article_control', 
         'time_last_click_article_treatment', 
         'click_count_article_control',
         'click_count_article_treatment'], axis=1, inplace=True)

#combine two dataframes together
mturk_df = mturk_df[['Answer.surveycode', 'WorkerId', 'AssignmentId', 'WorkTimeInSeconds']].copy()
mturk_df.columns = ['mturk_code', 'worker_id', 'assignment_id', 'task_time']
mturk_df['mturk_code'] = mturk_df.mturk_code.astype('str')

mturk_df.set_index('mturk_code', inplace=True)

df.set_index('mturk_code', inplace=True)

df = df.join(mturk_df)

df['phase'] = 'Primary'
```
